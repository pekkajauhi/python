```python
import requests
import pandas as pd
```

```python
# Eniten kannatusta saaneet kansalaisaloitteet

#url= 'https://www.kansalaisaloite.fi/api/v1/initiatives'

url='https://www.kansalaisaloite.fi/api/v1/initiatives?limit=50&orderBy=mostSupports'

r = requests.get(url)
print("Status code: ", r.status_code)

submissions = r.json()


aloitteet = []
for item in range(0,len(submissions)):

    aloitteet.append((submissions[item]['name']['fi'],submissions[item]['totalSupportCount'], submissions[item]['state'], submissions[item]['endDate']))
    
    

labels = ['name', 'supportcount', 'state', 'enddate']
df = pd.DataFrame.from_records(aloitteet, columns=labels)
df.set_index('name').sort_values('supportcount', ascending=False)
#df.sort_values('supportcount', ascending=False)
```
