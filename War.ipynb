```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
war_data = pd.read_csv("Inter-StateWarData_v4.0.csv")
```

```python
war_data.head()
```

```python
#Which countries have fought the most wars?
#This is very easy to find out:
```

```python
war_data['StateName'].value_counts()[0:20]
```

```python
#List of countries in alphabetical order:

sorted(war_data['StateName'].unique())
```

```python
#Number of countries
war_data['StateName'].nunique()
```

```python
#Bodycount for the top20 deadliest wars:
war_deaths = war_data.groupby('WarName')['BatDeath'].sum()
```

```python
war_deaths.sort_values(ascending=False)[:20]
```

```python
deadlywars = war_deaths.sort_values(ascending=False)[0:10]
```

```python
#Barplot of the top10 deadliest wars:
deadlywars.plot('bar', figsize=(15,8), title="Deadliest wars in history")
plt.show()
```

```python
#Function which can be used to plot the war deaths for a specific country

def deaths_for_state(state):
    war_data[war_data['StateName']==state].groupby('WarName')['BatDeath'].sum().sort_values(ascending=False).plot('bar', figsize=(15,8), title="Deadliest wars: "+state)
    plt.show()
```

```python
deaths_for_state('China')
deaths_for_state('United States of America')
```

```python
deaths_for_state('United Kingdom')
```

```python
deaths_for_state("USSR")
```

```python
deaths_for_state('Russia')
```
